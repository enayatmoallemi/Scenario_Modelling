```python
# Replace the original name of outputs with a shortened version for plotting
ylabels = {'Total Population Indicator':'Total Population' ,
             'Total Primary Education Graduates Indicator':'Primary Education Graduates',
             'Total Secondary Education Graduates Indicator':'Secondary Education Graduates',
              'Total Tertiary Education Graduates Indicator':'Tertiary Education Graduates', 
              'GWP per Capita Indicator':'GWP per Capita',
             'Energy Demand Indicator': 'Energy Demand',
             'Coal Production Indicator':'Coal Production', 
              'Gas Production Indicator':'Gas Production', 
              'Oil Production Indicator':'Oil Production', 
              'Solar Energy Production Indicator':'Solar Energy Production', 
              'Wind Energy Production Indicator':'Wind Energy Production', 
            'Biomass Energy Production Indicator':'Biomass Energy Production',
             'Nonenergy Crops Production Indicator':'Nonenergy Crops Production', 
              'Livestock Production Indicator':'Livestock Production',  
              'Total Croplands Indicator':'Total Croplands', 
             'Forest Land Indicator':'Forest Land', 
              'Pasture Land Indicator':'Pasture Land', 
           'Urban Industrial Land Indicator' :  'Urban Industrial Land',
              'CO2 Radiative Forcing Indicator': 'CO2 Radiative Forcing',
             'Total CO2 Emissions Indicator': 'Total CO2 Emissions'}
```

```python
# Count how many lines we will have in the top 10 parameters across outcomes for assigning linestyles and colors in legends 
new_mu_df = pd.DataFrame()
t = 2100
sc = 5000 # Specify the highest number of scenarios you ran
outcome_vars = list(ylabels.keys())
sa_dir = 'C:/Users/moallemie/GitHub/Scenario_Modelling/Results/Morris_results/'
for i, outcome_var in enumerate(outcome_vars):    
    mu_df = pd.read_csv(sa_dir+"MorrisIndices_{}_sc{}_t{}_test.csv".format(outcome_var, sc, t))
    mu_df.set_index('Unnamed: 0', inplace=True)
    mu_df.index.names = ['Uncertainty']
    mu_df.rename(columns={'mu_star': outcome_var}, inplace=True)
    mu_df.drop(['mu_star_conf', 'mu', 'sigma'], axis=1, inplace=True)
    mu_df.sort_values(by=[outcome_var], ascending=False, inplace=True)
    mu_df2 = mu_df.head(10)
    new_mu_df = pd.concat([new_mu_df, mu_df2], axis=1)
```

```python
# Create a dictionary for linestyle and linecolor for the figure legend.
color_list = sns.color_palette()

df_indices = [item[:-10] for item in new_mu_df.index.tolist()]
line_keys = {}
for p, par in enumerate(df_indices):
    if p < 10:
        line_keys[par] = ['-', color_list[p]]
    if p>=10 and p<20:
        line_keys[par] = ['--', color_list[p-10]]
    if p>=20 and p<30:
        line_keys[par] = ['-.', color_list[p-20]]
    if p>=30 and p<40:
        line_keys[par] = [':', color_list[p-30]]
    if p>=40 and p<50:
        line_keys[par] = [(0, (3, 5, 1, 5, 1, 5)), color_list[p-40]]
    
```

```python
from textwrap import wrap
from matplotlib.lines import Line2D


outcome_vars = list(ylabels.keys())

sa_dir='C:/Users/moallemie/GitHub/Scenario_Modelling/Results/Morris_results/'

color_list = sns.color_palette()

fig, axes = plt.subplots(5, 4, figsize=(18, 15), sharex='col')
fig.subplots_adjust(hspace=0.15, wspace=0.15)


for v, outcome_var in enumerate(outcome_vars): 
    df = pd.DataFrame(columns=[250, 500, 750, 1000, 1500, 2000, 3000, 5000])
    t = t
    dic = {}
    for sc in [250, 500, 750, 1000, 1500, 2000, 3000, 5000]:
        dic[sc] = pd.read_csv(sa_dir+"MorrisIndices_{}_sc{}_t{}_test.csv".format(outcome_var, sc, t), index_col=0)
        df[sc] = dic[sc]['mu_star']
    
    df.sort_values(by=[5000], ascending=False, inplace=True)
    df = df.head(10)
    
    # To remove Variation from the end of uncertain parameters in the final plots
    idx = [item for item in df.index]
    new_idx = [item[:-10] for item in df.index]
    df = df.rename(index=dict(zip(idx,new_idx))) 

    if v < 5:
        x = v
        y = 0
    elif v>=5 and v<10:
        x = v-5
        y = 1
    elif v>=10 and v<15:
        x = v-10
        y = 2
    elif v>=15 and v<20:
        x = v-15
        y = 3
    
    for i, par in enumerate(df.index):
        if i < 10:
            df.iloc[i, :].plot(ax=axes[x,y], linestyle=line_keys[par][0], marker='o', color=line_keys[par][1], linewidth=1, legend=False, fontsize=8)
        elif i>=10 and i<20:
            df.iloc[i, :].plot(ax=axes[x,y], linestyle=line_keys[par][0], color=line_keys[par][1], linewidth=1, legend=False, fontsize=8)
        elif i>=20 and i<30:
            df.iloc[i, :].plot(ax=axes[x,y], linestyle=line_keys[par][0], color=line_keys[par][1], linewidth=1, legend=False, fontsize=8)
        elif i>=30 :
            df.iloc[i, :].plot(ax=axes[x,y], linestyle=line_keys[par][0], color=line_keys[par][1], linewidth=1, legend=False, fontsize=8)



    axes[x,y].set_xticks([250, 500, 750, 1000, 1500, 2000, 3000, 5000])
    # Calculate the number of Morris experiments scenarios * (parameters+1)
    axes[x,y].set_xticklabels([250*115, 500*115, 750*115, 1000*115, 1500*115, 2000*115, 3000*115, 5000*115], 
                              rotation=90, fontsize=9, ha='center')
    axes[x,y].set_xlabel("Number of experiments (N)", fontsize=10)
    axes[x,0].set_ylabel("Morris mu star", fontsize=10)
    axes[x,y].annotate('\n'.join(wrap(ylabels[outcome_var], 30)), xy=(.5, 1.01), fontsize=10, xycoords='axes fraction', ha='center')
    #ax[x,y].legend().set_visible(False)


fig.subplots_adjust(bottom=0.2) # <-- Change the 0.2 to work for your plot.

legend_elements = []
for p, par in enumerate(df_indices):
    legend_elements = legend_elements + [Line2D([0], [0], linestyle=line_keys[par][0], color=line_keys[par][1], marker="o", lw=2, label=par)]
fig.legend(handles=legend_elements, loc='lower center', ncol=4, fontsize=8.5, bbox_to_anchor=(0.45, .01), fancybox=True, borderaxespad=0)



fig_dir='C:/Users/moallemie/GitHub/Scenario_Modelling/Fig/'
#plt.savefig(fig_dir+"Morris_covergence.pdf", dpi=600, bbox_inches='tight')
#plt.savefig(fig_dir+"Morris_covergence.png", dpi=600, bbox_inches='tight')  
#plt.savefig(fig_dir+"Morris_covergence_300dpi.png", dpi=300, bbox_inches='tight')    
```
