```python
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns
from matplotlib.patches import Rectangle
from textwrap import wrap
import matplotlib.image as mpimg

# Specify the range of variables to show in 5 subplots. This is used to make  5 subplots fitting into one A4 size page.
# The ranges should be [0,5], [5,10], [10,15], [15,20] for 20 outcomes
var_range = [10, 15]


ylabels = {'Total Population Indicator':'Total Population' ,
             'Total Primary Education Graduates Indicator':'Primary Education Graduates',
             'Total Secondary Education Graduates Indicator':'Secondary Education Graduates',
              'Total Tertiary Education Graduates Indicator':'Tertiary Education Graduates', 
              'GWP per Capita Indicator':'GWP per Capita',
             'Energy Demand Indicator': 'Energy Demand',
             'Coal Production Indicator':'Coal Production', 
              'Gas Production Indicator':'Gas Production', 
              'Oil Production Indicator':'Oil Production', 
              'Solar Energy Production Indicator':'Solar Energy Production', 
              'Wind Energy Production Indicator':'Wind Energy Production', 
            'Biomass Energy Production Indicator':'Biomass Energy Production',
             'Nonenergy Crops Production Indicator':'Nonenergy Crops Production', 
              'Livestock Production Indicator':'Livestock Production',  
              'Total Croplands Indicator':'Total Croplands', 
             'Forest Land Indicator':'Forest Land', 
              'Pasture Land Indicator':'Pasture Land', 
           'Urban Industrial Land Indicator' :  'Urban Industrial Land',
              'CO2 Radiative Forcing Indicator': 'CO2 Radiative Forcing',
             'Total CO2 Emissions Indicator': 'Total CO2 Emissions'}


fig, axes = plt.subplots(5, 2, figsize=(25, 29),gridspec_kw={'width_ratios': [1, .7]}, sharex='col')
fig.subplots_adjust(hspace=0.05, wspace=0.01)
sns.set(font_scale=1.2)
sns.set_style(style='white')

sa_dir='C:/Users/moallemie/EM_analysis/Data/'
img_dir = 'C:/Users/moallemie/EM_analysis/Fig/sa_verification/cropped-images/'

unc_temp = []

for i, outcome_var in enumerate(outcome_vars[var_range[0]:var_range[1]]):
    mu_df = pd.read_csv(sa_dir+"MorrisIndices_{}_sc{}_t{}.csv".format(outcome_var, sc, t))
    mu_df.set_index('Unnamed: 0', inplace=True)
    mu_df.index.names = ['Uncertainty']
    
    # Get the top 20 factors. This is for having a shared x axis with the combined top 20 list of uncertain parameters for each 5 output variables
    topfactor = list(mu_df.nlargest(20, 'mu_star').index)
    unc_temp = unc_temp + topfactor

    top_unc = list(set(unc_temp))

    mu_df = pd.read_csv(sa_dir+"MorrisIndices_{}_sc{}_t{}.csv".format(outcome_var, sc, t))
    mu_df.set_index('Unnamed: 0', inplace=True)
    mu_df.index.names = ['Uncertainty']
    mu_df2=mu_df[mu_df.index.isin(top_unc)]   

    
    # To remove Variation from the end of uncetain parameters in the final plots
    idx = [item for item in mu_df2.index]
    new_idx = [item[:-10] for item in mu_df2.index]
    mu_df2 = mu_df2.rename(index=dict(zip(idx,new_idx)))
    
    # Get the top n factors where the highest correlation is achieved
    n = top_factors[outcome_var] 
    topfactor2 = list(mu_df2.nlargest(n, 'mu_star').index) 
    
    # To get Morris index and errors
    ind = mu_df2.iloc[:,0]
    err = mu_df2.iloc[:,1]
    
    # Highlighting the topfactor bars with a differnet color
    colors = []
    for bar in ind.index.tolist(): 
        if bar in topfactor2:
            colors.append('#8DCCFC')
        else:
            colors.append('#FFE0D5')    
    
    axes[i,0].bar(x=ind.index.tolist(), height=ind.values.tolist(), yerr=err.values.T,color = colors, 
                  ecolor='dimgray', capsize=2,  width=.9)
    
    axes[i,0].annotate('\n'.join(wrap(ylabels[outcome_var], 30)), xy=(.5, .95), xycoords='axes fraction', ha='center')
    axes[i,0].margins(x=.01)
    axes[i,0].set_ylabel('mu_star index')

    img = mpimg.imread(img_dir+str(outcome_var)+'_'+str(top_factors[outcome_var])+'_topfactors.png')
    axes[i,1].imshow(X=img)
    axes[i,1].set_xticks([])
    axes[i,1].set_yticks([])
    plt.box(False)
    
    axes[i,1].spines['top'].set_visible(False)
    axes[i,1].spines['right'].set_visible(False)
    axes[i,1].spines['bottom'].set_visible(False)
    axes[i,1].spines['left'].set_visible(False)

    xlables = axes[i,0].get_xticklabels()
    axes[i,0].set_xticklabels(xlabels, rotation=90, ha='center', fontsize=14)

plt.savefig('{}/Morris_ranking_{}-{}_sc{}.png'.format(r'C:/Users/moallemie/EM_analysis/Fig/', var_range[0], var_range[1], sc), dpi=300,  bbox_inches='tight')
```
